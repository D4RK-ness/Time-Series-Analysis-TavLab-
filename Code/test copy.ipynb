### SARIMAX Model

Motivation: 
- SARIMAX allows for the incorporation of exogenous variables, also known as covariates or explanatory variables, into the model. Exogenous variables are variables that are not a part of the time series being modeled, but may have an impact on it. Here, we can introduce exogenous variables to simulate the waves of the pandemic and the effect of the lockdown on the number of cases.

- SARIMAX is a generalization of the SARIMA model that allows the specification of an autoregressive model with exogenous regressors. The SARIMAX model is written as:

$$y_t = c + \phi_1 y_{t-1} + \phi_2 y_{t-2} + \dots + \phi_p y_{t-p} + \theta_1 \epsilon_{t-1} + \theta_2 \epsilon_{t-2} + \dots + \theta_q \epsilon_{t-q} + \beta_1 x_{t-1} + \beta_2 x_{t-2} + \dots + \beta_r x_{t-r} + \epsilon_t$$

where $y_t$ is the time series being modeled, $x_t$ is the exogenous variable, $\epsilon_t$ is the error term, and $p$, $q$, and $r$ are the orders of the autoregressive, moving average, and exogenous regressor terms, respectively. The parameters $\phi_i$, $\theta_i$, and $\beta_i$ are the corresponding autoregressive, moving average, and exogenous regressor coefficients, respectively. The parameter $c$ is the intercept term.



```python
# Tuning the parameters for the SARIMAX model
import itertools
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
# Iterate through all the combinations of parameters, find the best combination and print the AIC value
import warnings
warnings.filterwarnings("ignore")

import statsmodels.api as sm

# Implementing the SARIMAX model
from statsmodels.tsa.statespace.sarimax import SARIMAX

# Split the data into train and test
from sklearn.model_selection import train_test_split

train, test = train_test_split(df_Delhi['Ratio'], test_size=0.2, shuffle=False)

# Fit the model

# Find the best combination of order
best_order = (0, 0, 0)
best_seasonal_order = (0, 0, 0, 0)

for param in pdq:
    for param_seasonal in seasonal_pdq:
        try:
            mod = SARIMAX(train, order=param, seasonal_order=param_seasonal, enforce_stationarity=False, enforce_invertibility=False)
            results = mod.fit()
            print('ARIMA{}x{}12 - AIC:{}'.format(param, param_seasonal, results.aic))
            if results.aic < best_aic:
                best_aic = results.aic
                best_order = param
                best_seasonal_order = param_seasonal
        except:
            continue

# Print the best combination of order
print('Best ARIMA{}x{}12 model - AIC:{}'.format(best_order, best_seasonal_order, best_aic))
```

As we can see, the best performance is achieved for

ARIMA(1, 0, 1)x(0, 0, 0, 12)12 - AIC:-28952.934841361173

ie, best order is (1,0,1) and best seasonal order is (0,0,0,12)

```python
# Fit the model with the best combination of order
best_order = (1, 0, 1)
best_seasonal_order = (0, 0, 0, 12)

# Fit the model
mod = SARIMAX(train, order=best_order, seasonal_order=best_seasonal_order, enforce_stationarity=False, enforce_invertibility=False)
results = mod.fit()

# Print the summary of the model
print(results.summary().tables[1])

# Plot the diagnostics of the model
results.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
# Predict the values for the test set
# Run SARIMAX on the test set
pred = SARIMAX(test, order=best_order, seasonal_order=best_seasonal_order, enforce_stationarity=False, enforce_invertibility=False).fit()

pred_ci = pred.conf_int()

# Plot the actual values and predicted values of the test set
ax = test.plot(label='observed', figsize=(15, 10))
plt.show()

pred.plot_diagnostics(figsize=(15, 12))
plt.show()
```

```python
# Plotting the forecasted values for the next 12 months
pred_uc = results.get_forecast(steps=12)
pred_ci = pred_uc.conf_int()

ax = df_Delhi['Ratio'].plot(label='observed', figsize=(15, 10))
pred_uc.predicted_mean.plot(ax=ax, label='Forecast')
ax.fill_between(pred_ci.index, pred_ci.iloc[:, 0], pred_ci.iloc[:, 1], color='k', alpha=.25)

ax.set_xlabel('Date')
# Format the Date as month-year using the DateFormatter
# ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%Y'))

ax.set_ylabel('Ratio')
plt.legend()
plt.show()
```
